# Seaborn

### The Titanic dataset was used to draw the graphs

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

df = sns.load_dataset('titanic')

df.head(3)
```

### Distribution charts

#### Dist Plot

```python
sns.displot(df['fare'] , bins = 10)
```

```python
sns.jointplot( x = 'age' , y = 'fare' , data = df)
```

```python
sns.jointplot( x = 'age' , y = 'fare' , data = df , kind= 'hex')
```

### Pair Plot

```python
dataset = df.dropna()  # All null items must be removed before execution
sns.pairplot(dataset)
```

### Rug

```python
sns.rugplot(dataset['fare'])
```

## 2 - Categorical Plots

### Bar Plot

```python
df = sns.load_dataset('titanic')
sns.barplot( x= 'sex' , y='age' , data=df)
```

```python
sns.barplot(x='sex', y='age', data=dataset, estimator=np.std)
```

Count Plot

```python
sns.countplot(x='sex' , data= df)
```

### Box Plot

```python
sns.boxenplot(x='sex' ,  y = 'age' , data= df )
```

```python
    sns.boxplot(x='sex', y='age', data=df, hue="survived")
```

### Violin Plot

```python
sns.violinplot( x= 'sex' , y='age' , data=df)
```

```python
    sns.violinplot(x='sex', y='age', data=df, hue='survived')
```

```python
    sns.violinplot(x='sex', y='age', data=df, hue='survived', split=True)
```

### Strip Plot

```python
sns.stripplot(x = 'sex' , y = 'age' , data=df )
```

```python
sns.stripplot(x = 'sex' , y = 'age' , data=df ,  hue='survived' , dodge=True) # split
```

### Swarm Plot

```python
    sns.swarmplot(x='sex', y='age', data=df)
```

```python
    sns.swarmplot(x='sex', y='age', data=dataset, hue='survived')
```

```python
    sns.swarmplot(x='sex', y='age', data=df, hue='survived', dodge=True)
```

```python
    sns.violinplot(x='sex', y='age', data=df)  
    sns.swarmplot(x='sex', y='age', data=df, color='black')
```

## Heat Maps

```python
corr = df.corr(numeric_only= True)
corr
```

```python
sns.heatmap(corr)
```

```python
sns.heatmap(corr, annot=True  )
```

```python
sns.heatmap(corr, annot=True ,cmap='winter')
```

```python
sns.heatmap(corr, annot=True , linecolor='blue', linewidth=1 )
```

## Facet Grids
